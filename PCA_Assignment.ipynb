```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Import the housing dataset and ensure it is imported properly

train = pd.read_csv("train.csv")
```

```python
train.head()
```

```python
# Check for missing values

percent_missing = 100 * train.isnull().sum()/len(train)
missing_value_df = pd.DataFrame({'column_name':train.columns,
                                'percent_missing':percent_missing})
```

```python
missing_value_df = missing_value_df[percent_missing>=40]
```

```python
missing_value_df = missing_value_df.sort_values("percent_missing", axis = 0, ascending = False)
```

```python
missing_value_df
```

```python
train.head(3)
```

```python
train = train.drop('Id',axis=1)
```

```python
train.head()
```

```python
# Keep columns with 60% or more valid data available 
df = train.dropna(how='any',axis=1,thresh=train.shape[0]*0.6)

```

```python
df.shape
```

```python
# 6 columns have been deleted.  These include those with more than 40% missing values and the Id column.
```

```python
df.head()
```

```python
df.info()
```

```python
# Select the numeric and categorical columns and check for the missing values

df_num = df.select_dtypes(exclude='object')
df_objs = df.select_dtypes(include='object')
```

```python
df_num_missing = df_num.isnull().sum()
df_num_missing = df_num_missing[df_num_missing>0]
df_num_missing
```

```python
df_objs_missing = df_objs.isnull().sum()
df_objs_missing = df_objs_missing[df_objs_missing>0]
df_objs_missing
```

```python
# Fill numeric missing columns with their median

num_cols = ['LotFrontage','MasVnrArea','GarageYrBlt']
df[num_cols] = df[num_cols].fillna('median')
```

```python
df[num_cols].isnull().sum()
```

```python
df_objs_missing.index
```

```python
cat_cols = ['MasVnrType', 'BsmtQual', 'BsmtCond', 'BsmtExposure', 'BsmtFinType1','BsmtFinType2', 'Electrical', 'GarageType', 'GarageFinish','GarageQual', 'GarageCond']
```

```python
df[cat_cols]=df[cat_cols].fillna(df.mode().iloc[0])
```

```python
df.isnull().sum()
```

```python
# Convert the categorical columns to dummy variable
```

```python
data = pd.get_dummies(df,drop_first=True)
```

```python
data.head()
```

```python
# Split the data into train and test sets
```

```python
X = data.drop('SalePrice',axis=1)
y = data['SalePrice']
```

```python
#Check the X dataframe
X.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.20,random_state=101)
```

```python
# Import linear regression algorithm

from sklearn.linear_model import LinearRegression
Lr = LinearRegression()
Lr.fit(X_train,y_train)
```

```python
y_pred = Lr.predict(X_test)
```

Find RMSE and R_Square

```python
# Evaluate the performance 

from sklearn.metrics import mean_squared_error,r2_score
```

```python
MSE = mean_squared_error(y_test,y_pred).round(2)
RMSE = np.sqrt(MSE).round(2)
```

```python
MSE
```

```python
RMSE
```

```python
rsquare = Lr.score(X_train,y_train).round(2)
```

```python
rsquare
```

```python
# Load the libraries

from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
# Standardize the features
scale = StandardScaler()
X_train_scale = scale.fit_transform(X_train)
```

```python
# Create a PCA that will retain 90% of variance

pca = PCA(n_components=0.99,whiten=True)
```

```python
# Conduct PCA

features_pca = pca.fit_transform(X_train_scale)
```

Number of Features in the Original and Reduced Dataset

```python
# Show results

print("Original number of features:",X_train_scale.shape[1])
print("Reduced number of features:",features_pca.shape[1])
```

Tranforming the test features

```python
X_test_scale = scale.transform(X_test)
```

Fit Linear Regression Model on the PCA transformed Data

```python
Lr.fit(X_train_scale,y_train)
```

```python
y_pred2 = Lr.predict(X_test_scale)
```

```python
MSE1 = mean_squared_error(y_test,y_pred2).round(2)
RMSE1 = np.sqrt(MSE1).round(2)
```

```python
MSE1
```

```python
RMSE1
```

```python
rsquare1 = Lr.score(X_train_scale,y_train).round(2)
```

```python
rsquare1
```

Applying Min-Max Scaler

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
#instantiate MinMaxScaler

scaler = MinMaxScaler()
```

```python
X_train_scaler = scaler.fit_transform(X_train)
```

```python
# Create Thresholder

from sklearn.feature_selection import VarianceThreshold

thresholder = VarianceThreshold(threshold=.1)
```

```python
# Create features with 0.1 matrix

created_features = thresholder.fit_transform(X_train_scaler)
```

```python
print("Original number of features:",X_train_scaler.shape[1])
print("Number of features with variance above 0.1 is equal to:",created_features.shape[1])
```

Transform test features

```python
X_test_scaler = scaler.transform(X_test)
```

Run a Linear Regression on Created Features

```python
Lr.fit(X_train_scaler,y_train)
```

```python
y_pred3 = Lr.predict(X_test_scaler)
```

```python
MSE2 = mean_squared_error(y_test,y_pred3).round(2)
RMSE2 = np.sqrt(MSE2).round(2)
```

```python
MSE2
```

```python
RMSE2
```

```python
rsquare2 = Lr.score(X_train_scaler,y_train).round(2)
```

```python
rsquare2
```

### Summarizing the Results

```python
R_Square_All_Features = rsquare
R_Square_PCA_Scaler = rsquare1
R_Square_MinMax = rsquare2
MSE_All_Features = MSE
MSE_PCA_Scaler = MSE1
MSE_MinMax = MSE2
RMSE_All_Features = RMSE
RMSE_PCA_Scaler =RMSE1
RMSE_MinMax = RMSE2
```

```python
summary = {'All_Features':[R_Square_All_Features,MSE_All_Features,RMSE_All_Features],
           'StandardScaler':[R_Square_PCA_Scaler,MSE_PCA_Scaler,RMSE_PCA_Scaler],
           'MinMax':[R_Square_MinMax,MSE_MinMax,RMSE_MinMax]}
          
Metrics = pd.DataFrame(summary,index=['R_square','MSE','RMSE'])

print(Metrics)
```

The results shows that when all the features were used for the training, the coefficient of determination was thesame with when the features were reduced through StandardScaler and MinMax.  This shows that using all the features for training was needless.  The MSE was much lower with MinMax scaler Reduction than when all the features were used and also compared with PCA standard scaler.  Similarly the RMSE was at its lowest with MinMax Scaler.  On the whole, it is better to perform feature reduction before training our models to optimize computing power.
